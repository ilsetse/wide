```python
# suggestion; look up all subjects in YSO

#%reset -f
import requests
import time


header = {'accept': 'application/json',
        'content-type': 'application/json'}

#### USER INPUT ####
search_query = "EEG"
####################
year = 2017
limit = 4#25

####################


# sort by descending date (2018, 2017, 2016)
url = ("https://api.finna.fi/api/v1/search?lookfor=" + search_query + 
       "&field[]=year&field[]=title&type=Subject&field[]=subjects" +
       "&filter[]=search_daterange_mv%3A%22%5B2017%20TO%202018%5D%22" + # 2017 - 2018
       #"&filter[]=search_daterange_mv%3A%22%5B2017%20TO%202017%5D%22" + # 2017 - 2017
       "&sort=main_date_str%20desc&page=1&prettyPrint=false&lng=fi")

# TODO is there a way to specify year in field?

t0 = time.time()

results = []
while True:
    response = requests.get(url, headers=header).json()
    results += response['records']

    resultCount = response['resultCount']
    
    
    if len(results) == resultCount or len(results) >= limit*20:
        break

t1 = time.time()

total = t1-t0
print("time: ", total)

print("result count: ", resultCount)
print("num entries: ", len(results)) # limit*20

numKeywords = 0
for i in range(resultCount):
    if results[i]['year'] == str(year):
        numKeywords +=1
    
numKeywords
```
